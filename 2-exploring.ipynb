### With each numerical column, how are values distributed?

```python
df.describe()
```

**What is the percentage of missing values?**

```python
num_df = df.select_dtypes(include=["number"])
def percent_missing(series): 
    return series.isnull().sum() * 100 / len(series)
print("The percentage of missing values in each columns:")
num_df.agg(percent_missing)
```

**Min? max? Are they abnormal?**

```python
num_df.agg([min, max])
```

Dữ liệu về tuổi có dấu hiệu bất thường: 
+ `max = 99999999999` thì không phù hợp với thực tế.
+ `min = -1726` là dữ liệu sai vì tuổi không tồn tại giá trị âm.
